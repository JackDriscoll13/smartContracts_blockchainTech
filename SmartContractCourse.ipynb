<span style="color:Green"> How it Works: </span> 
  
- [BlockChain 101 Video](https://www.youtube.com/watch?v=_160oMzblY8)
- [Visualising Blockchain Practice](https://www.andersbrownworth.com/blockchain)
- Hash: a unique fixed length streng. Meant to identify a piece of data. Created by putting data into a hash function (algorithm)
- The hash corresponds to the specific data in the block. ie, one block.
- Nonce: the nonce is the solution to the block, the number where the specifc data matches the corresponding hash. 
- NONCE + DATA (tx) = unique hash
- A node gets paid to "mine" or solve the block, by matching the nonce to the specific hash requirement (i.e. 4 zeros at the start of the hash) 
- Prev: points to the hash of the previous block
- Changing one piece of data invalidates the entire block chain
- Peer to peer: Each peer (node), each independent node has the same power, a mechanism of checks and balances for checking to make sure the blockchain is correct. Matching (hashes)
- TX: transaction data.

<span style="color:Green"> Consensus: </span>
- Consesus is the mechanism used to agree on the state of the blockchain. 
- Consessus is composed of 1. Chain selection   2. Sybil Resistance 
    - Sybil Reistance: A blockchains defense against people creating multiple identies (nodes) to have more power over the system
      - Proof of work (PoW): Mining - solving the complex problem. Very computationally expensive. chains can change the block time by changing difficulty of problem. 
        - Pros: Great sybil resistance. cons: energy expensive and enviormentally impactful
      - Proof of Stake (PoS): Staking ... Nodes put up colleteral as evidence that they will "behave" on the block. miners replaced by validators. Validators are stakers that are chosen randomly
      - to  solve a block and then validated by the rest of the nodes
        - Pros: great sybil resistance mechanism, uses less energy. Cons: slightly less decentralized because of cost of stake
    - Chain Selection Rule:
      - In nakamoto consesus, the longest chain wins.

<span style="color:Green"> Types of Attacks That Can Happen: </span>
- Sybil Attack: One user creates many nodes and uses lots of computational resources to gain unfair power. (less prevalent)
- 51% attack: The longest chain is the correct one so long as it matches up with 51% of the rest of the network. (more prevalent) (someone )
